```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style= 'darkgrid', 
       color_codes=True,
       font = 'DejaVu Sans',
       font_scale= 1.5)
```

### Read the data set

```python
train = pd.read_csv("dataset/train.csv")
test = pd.read_csv("dataset/test.csv")

print(f'The train and test data sets dimensions are {train.shape}, {test.shape}\n')

train.head()
```

```python
# Count the number of words per each comment
words_per_comments_train = train.Comment.map(lambda x : len(x.split()))

print(f'Longest comment is {words_per_comments_train.max()} word')
print(f'Shortest comment is {words_per_comments_train.min()} word\n')

# Show the properties of the data
plt.figure(figsize=(20,8))

# Plot a histogram of the number of words per each comment
plt.subplot(121)
plt.hist(words_per_comments_train, bins=100)
plt.title("Train: Distribution of words per comment")
plt.ylabel('frequency')
plt.xlabel("Word count")

# Plot a count plot for the comments per each topic
plt.subplot(122)
sns.countplot(x=train.Topic)
plt.title("Train: Count plot of Tpoics")
plt.show();
```

```python
# Count the number of words per each comment
words_per_comments_test = test.Comment.map(lambda x : len(x.split()))

print(f'Longest comment is {words_per_comments_test.max()} word')
print(f'Shortest comment is {words_per_comments_test.min()} word\n')

# Show the properties of the data
plt.figure(figsize=(20,8))

# Plot a histogram of the number of words per each comment
plt.subplot(121)
plt.hist(words_per_comments_test, bins=100)
plt.title("Test: Distribution of words per comment")
plt.ylabel('frequency')
plt.xlabel("Word count")

# Plot a count plot for the comments per each topic
plt.subplot(122)
sns.countplot(x=test.Topic)
plt.title("Test: Count plot of Tpoics")
plt.show();
```
